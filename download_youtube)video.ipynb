```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
website = requests.get("http://coreyms.com").text
```

```python
soup = BeautifulSoup(website , 'lxml')
```

```python
article = soup.find('article')
```

```python
vid = article.find('iframe', class_= "youtube-player").attrs['src']
```

```python
vid_id = vid.split('/')
```

```python
video = vid_id[-1].split('?')
```

```python
c = video[0]
```

```python
ytLink = f'https://youtube.com/watch?v={c}'
```

```python
from pytube import YouTube
```

```python
VIDEO = YouTube(ytLink)
```

```python
stream = VIDEO.streams.get_highest_resolution()
```

```python
stream.download('Desktop')
```
